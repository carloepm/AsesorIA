REGISTRO DE USUARIOS

```python
class SQLServerUsuarios:

    def __init__(self, conexion):
        self.conexion = conexion
        self.max_intentos = 6
        self.intentos = 0  

    def add_usuario(self ,NombreUsuario, CorreoUsuario, NombrePlan, Contraseña):
        self.conexion = conexion_sql_server()
        if self.conexion:
            cursor = self.conexion.cursor()
            try:
                cursor.execute("INSERT INTO Usuarios (NombreUsuario, CorreoUsuario, NombrePlan, Contraseña) VALUES (?, ?, ?, ?)", 
                                (NombreUsuario, CorreoUsuario, NombrePlan, Contraseña))
                self.conexion.commit()                
            except Exception as e:
                print(f"Error al agregar usuario: {e}")
            finally:
                cursor.close()
                self.conexion.close()
    
    def _verificar_nombre(self, NombreUsuario):
        self.conexion = conexion_sql_server()
        if self.conexion:
            cursor = self.conexion.cursor()
            try:
                cursor.execute("SELECT * FROM Usuarios WHERE NombreUsuario = ?", (NombreUsuario))
                if cursor.fetchone():                    
                    return True
                else: 
                    return False                            
            except Exception as e:
                print(f"Error al verificar nombre de usuario: {e}")
            finally:
                cursor.close()
                self.conexion.close()
                
    def _verificar_correo(self, CorreoUsuario):
        self.conexion = conexion_sql_server()
        if self.conexion:
            cursor = self.conexion.cursor()
            try:
                cursor.execute("SELECT * FROM Usuarios WHERE CorreoUsuario = ?", (CorreoUsuario))
                if cursor.fetchone():
                    return True
                else: 
                    return False
            except Exception as e:
                print(f"Error al verificar correo: {e}")
            finally:
                cursor.close()
                self.conexion.close()    
    
    def _verificar_plan(self, NombrePlan):
        self.conexion = conexion_sql_server()
        if self.conexion:
            cursor = self.conexion.cursor()
            try: 
                cursor.execute("SELECT COUNT(*) FROM Planes WHERE NombrePlan = ?", (NombrePlan))
                planes = cursor.fetchall()
                return planes
            except Exception as e:
                print(f"Error al obtener planes {e}")
                return None
            finally:
                cursor.close()
                self.conexion.close()
                 
    def _verificar_contraseña(self, Contraseña):
        if self.intentos >= self.max_intentos:
            return False
        
        if (len(Contraseña) < 8 or 
            not re.search("[a-zA-Z]", Contraseña) or 
            not re.search("[0-9]", Contraseña) or 
            not re.search("[!@#$%^&*()-_=+{};:,<.>]", Contraseña)):
            return False
        else: 
            return True
```

```python
sql_usuarios = SQLServerUsuarios(conexion)
print("REGISTRO DE USUARIO")


while True:
    # Solicitar y verificar el nombre de usuario
    NombreUsuario = str(input("Usuario: "))
    if sql_usuarios._verificar_nombre(NombreUsuario):
        print("Ingrese un nombre de usuario diferente.")
    else:
        print("Usuario correcto.")
        break  # Salir del bucle si el usuario es válido

while True:
    # Solicitar y verificar el correo electrónico
    CorreoUsuario = str(input("Correo electrónico: "))
    if sql_usuarios._verificar_correo(CorreoUsuario):
        print("Correo electrónico existente.")
    else:
        print("Correo válido.")
        break  # Salir del bucle si el correo electrónico es válido

while True:
    # Solicitar y verificar el plan
    NombrePlan = str(input("Plan: "))
    if sql_usuarios._verificar_plan(NombrePlan):
        print("Plan correcto.")
        break  # Salir del bucle si el plan es válido
    else:
        print("Plan no disponible.")  


while True:
    # Solicitar y verificar la contraseña
    Contraseña = str(input("Contraseña: "))
    valid = sql_usuarios._verificar_contraseña(Contraseña)
    if valid == False:
        print("Contraseña incorrecta.")
    else:
        print("Contraseña correcta.")
        # Si todos los campos son válidos, agregar el asesor h y salir del bucle
        sql_usuarios.add_usuario(NombreUsuario, CorreoUsuario, NombrePlan, Contraseña)
        print("Registro exitoso.")
        break
```

REGISTRO ASESOR HUMANO

```python
class SQLServerAsesorH:
    
    def __init__(self, conexion):
        self.conexion = conexion
        self.max_intentos = 6
        self.intentos = 0  
        self.max_dni_digitos = 8
        
    def _add_asesor_h(self, dni, NombreAsesorH, ApellidoAsesorH, CorreoAsesorH, Especialidad, Credencial, Contraseña):
        self.conexion = conexion_sql_server()
        if self.conexion:
            cursor = self.conexion.cursor()
            try:
                    cursor.execute("INSERT INTO Asesor_H (dni, NombreAsesorH, ApellidoAsesorH, CorreoAsesorH, Especialidad, Credencial, Contraseña) VALUES (?, ?, ?, ?, ?, ?, ?)", (dni, NombreAsesorH, ApellidoAsesorH, CorreoAsesorH, Especialidad, Credencial, Contraseña))
                    self.conexion.commit()
            except Exception as e:
                    print(f"Error al agregar usuario {e}")
            finally:
                    cursor.close()
                    self.conexion.close()
    
    def _verificar_dni(self, dni):
        self.conexion = conexion_sql_server()
        if not re.match("^\d+$", dni) or len(dni) != self.max_dni_digitos:
            return False
        
        if self.conexion:
            cursor = self.conexion.cursor()
            try:
                cursor.execute("SELECT * FROM Asesor_H WHERE dni = ?", (dni))
                if cursor.fetchone():
                    return True
                else: 
                    return False
            except Exception as e:
                print(f"Error al verificar el dni: {e}")
            finally:
                cursor.close()
                self.conexion.close()
                
        
    def _verificar_correo(self, CorreoAsesorH):
        self.conexion = conexion_sql_server()
        if self.conexion:
            cursor = self.conexion.cursor()
            try:
                cursor.execute("SELECT * FROM Asesor_H WHERE CorreoAsesorH = ?", (CorreoAsesorH))
                if cursor.fetchone():
                    return True
                else:
                    return False
            except Exception as e:
                print(f"Error al verificar correo: {e}")
            finally:
                cursor.close()
                self.conexion.close()
         
    def _verificar_especialidad(self, Especialidad):
        self.conexion = conexion_sql_server()
        cursor = self.conexion.cursor() 
        cursor.execute("SELECT * FROM Especialidades WHERE Especialidad = ?", (Especialidad))
        if cursor.fetchone():
            self.conexion.commit()
            cursor.close()
            self.conexion.close()
            return True
            
        else:
            return False
 
    def _verificar_credencial(self, Credencial):
        query_credencial = """
            SELECT Credencial FROM Especialidades WHERE Credencial = ?;
                """
        with conexion.cursor() as cursor:
            cursor.execute(query_credencial, (Credencial))
            if cursor.fetchone():
                return True
            else:
                return False
 
    def _lista_credencialidades(self):
        self.conexion = conexion_sql_server()
        query_lista_crendencialidades = """SELECT Especialidad, Credencial FROM Especialidades"""
        with conexion.cursor() as cursor:
            cursor.execute(query_lista_crendencialidades)
            lista_credenciales = cursor.fetchall()
        return lista_credenciales
     

    def _verificar_contraseña(self, Contraseña):
        if self.intentos >= self.max_intentos:
            return False
        
        if (len(Contraseña) < 8 or
            not re.search("[a-zA-Z]", Contraseña) or
            not re.search("[0-9]", Contraseña) or
            not re.search("[!@#$%^&*()-_=+{};:,<.>]", Contraseña)):
            return False
        else:
            return True
                     
                
```

```python
from datetime import datetime

class SQLServerTransacciones:
    
    def __init__(self, conexion):
        self.conexion = conexion
        self.medios_pago_aceptados = "Efectivo, Transferencias, Naranja, Visa, Mastercard"
        
        
    def add_transaccion(self, FechaTransaccion, NombrePlan, MedioPago, Precio, ID_Usuario, ID_Plan):
        self.conexion = conexion_sql_server()
        if self.conexion:
            cursor = self.conexion.cursor()
            try:
                cursor.execute("INSERT INTO Transacciones (FechaTransaccion, NombrePlan, MedioPago, Precio, ID_Usuario, ID_Plan) VALUES (?, ?, ?, ?, ?, ?)", (FechaTransaccion, NombrePlan, MedioPago, Precio, ID_Usuario, ID_Plan))
                self.conexion.commit()
            except Exception as e:
                print(f"Error en la transaccion {e}")
            finally:
                cursor.close()
                self.conexion.close()
                
                
    def obtener_id_usuario(self, NombreUsuario):
        self.conexion = conexion_sql_server()
        if self.conexion:
            cursor = self.conexion.cursor()
            try:
                cursor.execute("SELECT ID_Usuario FROM Usuarios WHERE NombreUsuario = ?", (NombreUsuario))
                row = cursor.fetchone()
                if row:
                    return row[0] 
                else:
                    return None
            except Exception as e:
                print(f"Error al obtener ID del Usuario {e}")
            finally:
                cursor.close()
                self.conexion.close()    
        return None
                
                
    def obtener_id_plan(self, NombrePlan):
        self.conexion = conexion_sql_server()
        if self.conexion:
            cursor = self.conexion.cursor()
            try:
                cursor.execute("SELECT ID_Plan FROM Planes WHERE NombrePlan = ?", (NombrePlan))
                row = cursor.fetchone()
                if row:
                    return row[0]
                else:
                    return None
            except Exception as e:
                print(f"Error al obtener ID del Plan {e}")
            finally:
                cursor.close()
                self.conexion.close()    
        return None
        
      
    def verificar_fecha_pago(self, FechaTransaccion):
        try:
            fecha_formateada = FechaTransaccion.strftime('%d-%m-%Y')
            if fecha_formateada == datetime.strftime(FechaTransaccion, '%d-%m-%Y'):
                return True
            else:
                return False
        except AttributeError:
            return False
        
    
    def verificar_medio_pago(self, MedioPago):
        if MedioPago in self.medios_pago_aceptados:
            print(f"el medio de pago {MedioPago} es valido")
            return True
        else:
            return False
        
    
    def verificar_precio(self):
        query_precio = """
            SELECT NombrePlan, Precio FROM [Planes];
                """
        with conexion.cursor() as cursor:
            cursor.execute(query_precio)
            lista_precios = cursor.fetchall()
        return lista_precios
    
```

```python
sql_transacciones = SQLServerTransacciones(conexion)
print("REGISTRO DE TRANSACCIONES")
while True:
    NombreUsuario = str(input("Ingrese su nombre de usuario: "))
    idUsuario = sql_transacciones.obtener_id_usuario(NombreUsuario)
    if sql_transacciones.obtener_id_usuario(NombreUsuario):
        print("Usuario correcto")
        break
    else:
        print("Ha ingresa un usuario incorrecto o no existente.")

while True:
    NombrePlan = str(input("Plan: "))
    idPlan = sql_transacciones.obtener_id_plan(NombrePlan)
    if sql_transacciones.obtener_id_plan(NombrePlan):
        print("Plan disponible")
        break
    else:
        print("Seleccione un plan disponible dentro de Compras")

while True:
    FechaTransaccion_str = input("Fecha en formato DD-MM-YYYY: ")
    try:
        FechaTransaccion = datetime.strptime(FechaTransaccion_str, '%d-%m-%Y').date()
        if sql_transacciones.verificar_fecha_pago(FechaTransaccion):
            print("Fecha correcta.")
            break
        else:
            print("Formato de fecha incorrecto, debe ser día-mes-año.")
    except ValueError:
        print("Formato de fecha incorrecto, asegúrese de ingresar la fecha en formato DD-MM-YYYY.")

while True:
    MedioPago = str(input("Medio de pago: "))
    if sql_transacciones.verificar_medio_pago(MedioPago):
        print("Medio de pago aceptado.")
        break
    else:
        print("Medio de pago no disponible de momento.")

pago_coincidente = False  # Variable para seguir la coincidencia del precio con el plan correspondiente.

while not pago_coincidente:
    #El usuario no ingresa los precios, sino que los selecciona del plan correspindiente.
    #Es solo prueba para la clase que ingresa a la base de datos.
    Precio = float(input("Precio: "))
    lista_precios = sql_transacciones.verificar_precio()
    
    for nombreplan, precio in lista_precios:
        if nombreplan == NombrePlan and precio == Precio:
            pago_coincidente = True
            print("Pago valido.")
            sql_transacciones.add_transaccion(FechaTransaccion, NombrePlan, MedioPago, Precio, idUsuario, idPlan)
            print("Transaccion exitosa.")
            break  # Salir del bucle si se encuentra una coincidencia
    
    if not pago_coincidente:
        print("Pago no correspondiente.")
```
